```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -------------------------------
# Parameters
# -------------------------------
n_trials = 20        # Number of trials per experiment
p_success = 0.5      # Probability of success per trial
experiment_counts = [10, 100, 1000, 10000]  # Increasing experiment sizes

# -------------------------------
# Plotting
# -------------------------------
plt.figure(figsize=(14, 10))

for i, num_experiments in enumerate(experiment_counts, 1):
    # Simulate binomial experiments
    samples = np.random.binomial(n=n_trials, p=p_success, size=num_experiments)
    
    # Calculate theoretical normal distribution
    mean = n_trials * p_success
    std = np.sqrt(n_trials * p_success * (1 - p_success))
    x = np.linspace(0, n_trials, 1000)
    y = norm.pdf(x, mean, std)

    # Plot histogram
    plt.subplot(2, 2, i)
    plt.hist(samples, bins=range(n_trials+2), density=True, alpha=0.6,
             color='skyblue', edgecolor='black', label='Binomial Histogram')

    # Plot normal curve
    plt.plot(x, y, 'r--', label='Normal Approximation')

    plt.title(f'{num_experiments} Experiments')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()

plt.suptitle('Convergence of Binomial to Normal with Increasing Experiments', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
